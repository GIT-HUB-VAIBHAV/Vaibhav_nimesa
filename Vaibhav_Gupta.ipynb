```python
import requests

# Global variable for the API URL
API_URL = "https://samples.openweathermap.org/data/2.5/forecast/hourly"  # Replace with the actual API URL
API_KEY = "b6907d289e10d714a6e88b30761fae22"  # Replace with your actual API key

def get_weather_data(city, api_key, date):
    # Make a GET request to the API with the city and API key parameters
    response = requests.get(API_URL, params={"q": city, "appid": api_key})
    
    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        return extract_weather_data(data, date)
    else:
        print("Error fetching weather data.")
        return None

def extract_weather_data(data, date):
    # Extract weather data for the given date
    for item in data["list"]:
        if item["dt_txt"].startswith(date):
            return item
    return None

def main():
    city = "London,us"  # Change this to the desired city or location
    api_key = API_KEY

    while True:
        print("\nOptions:")
        print("1. Get weather")
        print("2. Get Wind Speed")
        print("3. Get Pressure")
        print("0. Exit")
        
        choice = input("Enter your choice: ")

        if choice == "1":
            date = input("Enter the date (YYYY-MM-DD): ")
            weather_data = get_weather_data(city, api_key, date)
            if weather_data:
                temperature = weather_data["main"]["temp"]
                print(f"Temperature on {date}: {temperature:.2f} °K")
        
        elif choice == "2":
            date = input("Enter the date (YYYY-MM-DD): ")
            weather_data = get_weather_data(city, api_key, date)
            if weather_data:
                wind_speed = weather_data["wind"]["speed"]
                print(f"Wind Speed on {date}: {wind_speed} m/s")

        elif choice == "3":
            date = input("Enter the date (YYYY-MM-DD): ")
            weather_data = get_weather_data(city, api_key, date)
            if weather_data:
                pressure = weather_data["main"]["pressure"]
                print(f"Pressure on {date}: {pressure:.2f} hPa")
        
        elif choice == "0":
            print("Exiting the program.")
            break

        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
